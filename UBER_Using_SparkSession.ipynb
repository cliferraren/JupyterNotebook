#### Import SparkSession

```scala
import org.apache.spark.sql.SparkSession
```

```scala
val spark = SparkSession.
            builder().
            appName("Uber Pickups").
            config("spark.sql.warehouse.dir",  "tmp/sparksql").
            getOrCreate()
            
```

```scala
val uber = spark.read.
                    option("header", true).
                    option("inferSchema", true).
                    csv("data/uber.csv")
```

#### Extract Features

```scala
val featureCols = Array("Lat" , "Lon")
```

```scala
import org.apache.spark.ml.feature.VectorAssembler
import org.apache.spark.ml.clustering.KMeans
```

#### Prepare Feature Vectors

```scala
val assembler = new VectorAssembler().
                                    setInputCols(featureCols).
                                    setOutputCol("Features")
                                    
val df2 = assembler.transform(uber)
```

```scala
df2.show(10)
```

#### Divide Data to Train and Test

```scala
val Array(trainData, testData) = df2.randomSplit(Array(0.7,0.3),5043)
```

```scala
println("Total number of records: "+ df2.count())
println("Train Data: "+ trainData.count())
println("Test Data: "+ testData.count())
```

#### Create Kmeans Model

```scala
val kmeans = new KMeans().setK(20).setFeaturesCol("Features").setMaxIter(5)
```

#### Apply Algorithm to Training Data - Prepare Model

```scala
val model = kmeans.fit(trainData)
```

#### Show Cluster Centers

```scala
model.clusterCenters.foreach(println)
```

#### Check the Cluster for Test Data

```scala
val category = model.transform(testData)
```

```scala
category.show(10)
```

#### Register DF as a Temp Table

```scala
category.createOrReplaceTempView("Uber")
```

#### How many pick-ups occurred in each cluster?

```scala
spark.sql("SELECT prediction AS LOCATION_CATEGORY, COUNT(prediction) AS NO_OF_TRIPS FROM Uber GROUP BY LOCATION_CATEGORY ORDER BY NO_OF_TRIPS DESC").show()
```
